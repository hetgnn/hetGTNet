```python
class config:
    dataset = 'ACM'
    data_path = '../GTN/data/'
    epoch = 500
    patience = 100
    n_hid = 64
    edge_dim = 64
    num_heads = 8
    dropout = 0.5
    dropout2 = 0.5
    lr = 0.01
    weight_decay = 0
    hop = 2
    negative_slope = 0.05
    beta = 0.05
    num_test = 30
    filter_pct = 0.1 # remove the top and bottom filer_pct points before obtaining statistics of test accuracy
    average = 'macro' # 'f1 average: can choose either macro or micro.'
    log_step = 1000 # training log step
    
args = config()
```
