```python
class config:
    embedder = 'DMGI'
    dataset = 'imdb'
    data_path = '../DMGI/data/'
    save_path = '../DMGI/saved_model/' # path for saving models
    metapaths = 'MAM,MDM'
    nb_epochs = 10000
    hid_units = 64
    lr = 0.0005
    l2_coef = 0.0001
    drop_prob = 0.5
    reg_coef = 0.001
    sup_coef = 0.1
    sc = 3.0
    margin = 0.1
    gpu_num = 0 # cuda id#
    patience = 20
    patience2 = 100 # patience for downstream task
    nheads = 1
    activation = 'relu'
    isSemi = False
    isBias = False
    isAttn = False
    num_iter = 500 # Max epochs to run for downstream task
    num_test = 30
    filter_pct = 0.1 # remove the top and bottom filer_pct points before obtaining statistics of test accuracy
    
args = config()
```

```python
datasets = ['acm', 'imdb', 'dblp']
for dataset in datasets:
    args.dataset = dataset
    if dataset == 'acm':
        args.metapaths = 'PAP,PSP'
        print('\nstart testing on ' + dataset + ' dataset')
        main(args)
    elif dataset == 'imdb':
        args.metapaths = 'MAM,MDM'
        print('\nstart testing on ' + dataset + ' dataset')
        main(args)
    elif dataset == 'dblp':
        args.metapaths = 'APA,APCPA,APTPA'
        print('\nstart testing on ' + dataset + ' dataset')
        main(args)
```
