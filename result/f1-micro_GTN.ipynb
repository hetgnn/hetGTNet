```python
root_dir = '../GTN' # dir of the source code
import sys
sys.path.append(root_dir)
from main_sparse import main
#from main import main
import numpy as np
```

```python
class config:
    dataset = 'ACM'
    data_path = '../GTN/data/'
    epoch = 500
    patience = 100
    node_dim = 32
    num_channels = 2
    lr = 0.005
    weight_decay = 0.001
    num_layers = 2
    adaptive_lr = 'true'
    num_test = 30
    norm = 'true'
    filter_pct = 0.1 # remove the top and bottom filer_pct points before obtaining statistics of test accuracy
    average = 'micro' # 'f1 average: can choose either macro or micro.'
    log_step = 1000 # training log step
    use_cpu = False
    
args = config()
```

```python
f1s = []
datasets = ['ACM', 'IMDB', 'DBLP']
for dataset in datasets:
    args.dataset = dataset
    if dataset == 'ACM':
        args.node_dim = 32
        args.num_layers = 2
        args.adaptive_lr = 'true'
        args.use_cpu = False
        print('\nstart testing on ' + dataset + ' dataset')
        f1s.append(main(args))
    elif dataset == 'IMDB':
        args.node_dim = 32
        args.num_layers = 3
        args.adaptive_lr = 'true'
        args.use_cpu = False
        print('\nstart testing on ' + dataset + ' dataset')
        f1s.append(main(args))
    elif dataset == 'DBLP':
        args.node_dim = 16
        args.num_layers = 2
        args.adaptive_lr = 'false'
        args.use_cpu = False
        print('\nstart testing on ' + dataset + ' dataset')
        f1s.append(main(args))
```

```python
f1s = np.array(f1s)
np.savetxt('f1-micro_GTN.txt',f1s)
```
